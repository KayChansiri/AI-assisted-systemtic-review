```python
import pandas as pd

# Load the dataset
df = pd.read_excel("Phase1_with_humans.xlsx")

# Define secondary study terms to search for
secondary_terms = ["meta-analysis", "systematic review", "review", "literature review"]

# Function to check if any secondary term is in the abstract
def contains_secondary_terms(abstract):
    if pd.isna(abstract):  # Handle missing abstracts
        return False
    abstract = abstract.lower()  # Case-insensitive search
    return any(term in abstract for term in secondary_terms)

# Add a new column to flag rows with secondary terms
df["contains_secondary_terms"] = df["abstract"].apply(contains_secondary_terms)

# Filter and display rows with potential secondary studies for manual review
secondary_rows = df[df["contains_secondary_terms"] == True][["abstract", "is.included"]]
print(f"\nFound {len(secondary_rows)} rows with potential secondary study terms:")
print(secondary_rows.to_string(index=True))

# Save the updated dataset with the new flag column
output_path = "Phase1_with_secondary_flags.xlsx"
df.to_excel(output_path, index=False)
print(f"\n✅ Dataset with flagged rows saved to: {output_path}")
print("Please manually review the 'contains_secondary_terms' column and update 'is.included' as needed.")
```
